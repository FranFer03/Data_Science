# **Cuarta Entrega**
Este conjunto de datos proporcionado por una organización de aerolíneas, que llamaremos "Aerolíneas Invistico”, está compuesto por 23 columnas y 130 mil columnas.
El conjunto de datos se compone de los datos de los clientes que ya han volado con ellos. Se han consolidado los comentarios de los clientes sobre diversos contextos y sus datos de vuelo.

### **Hipótesis:**

El objetivo principal es predecir si un futuro cliente estará satisfecho con el servicio, utilizando datos de parámetros anteriores.
Además, la aerolínea busca identificar áreas en las que deba enfocarse para mejorar y generar clientes más satisfechos.

### **Diccionario de datos**

* satisfaction: Representa el nivel de satisfacción del cliente con su
experiencia de vuelo en general. Es una variable categórica con dos posibles valores: "satisfied" o "dissatisfied".

* Gender: Indica el género del cliente. Es una variable categórica con dos posibles valores: "Male" o "Female".

* Customer Type: Categoriza a los clientes según su tipo. Es una variable categórica con dos posibles valores: "Loyal Customer" o "Disloyal Customer"

* Age: Representa la edad del cliente. Es una variable numérica expresada en años.

* Type of Travel: Indica el propósito del viaje del cliente. Es una variable categórica con dos posibles valores: "Business travel" o "Personal travel".

* Class: representa la clase de servicio que el cliente eligió. Es una variable categórica con tres posibles valores: "Eco", "Eco Plus", o "Business"

* Flight Distance: Esta columna denota la distancia del vuelo en millas. Es una variable numérica.

* Seat comfort: Esta columna representa la satisfacción del cliente con la comodidad del asiento. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Departure/Arrival time convenient: Esta columna denota la
satisfacción del cliente con la conveniencia de los horarios de salida y llegada. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Food and drink: Esta columna representa la satisfacción del cliente con la comida y bebida proporcionada durante el vuelo. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Gate location: Esta columna denota la satisfacción del cliente con la ubicación de la puerta de embarque. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Inflight wifi service: Esta columna representa la satisfacción del cliente con el servicio de wifi a bordo. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Inflight entertainment: Esta columna denota la satisfacción del cliente con las opciones de entretenimiento a bordo. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Online support: Esta columna representa la satisfacción del cliente con el soporte en línea. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Ease of Online booking: Esta columna denota la satisfacción del cliente con la facilidad de reserva en línea. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* On-board service: Esta columna representa la satisfacción del cliente con el servicio a bordo proporcionado por la aerolínea. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Leg room service: Esta columna denota la satisfacción del cliente con el espacio para las piernas proporcionado durante el vuelo. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Baggage handling: Esta columna representa la satisfacción del cliente con el manejo del equipaje por parte de la aerolínea. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Checkin service: Esta columna denota la satisfacción del cliente con el servicio de check-in de la aerolínea. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Cleanliness: Esta columna representa la satisfacción del cliente con la limpieza de la aeronave. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Online boarding: Esta columna denota la satisfacción del cliente con el proceso de embarque en línea. Es una variable numérica, medida en una escala de 0 a 5, donde valores más altos indican mayor satisfacción.

* Departure Delay in Minutes: Esta columna representa la duración del retraso en la salida en minutos para cada vuelo. Es una variable numérica.
* Arrival Delay in Minutes: Esta columna denota la duración del retraso en la llegada en minutos para cada vuelo. Es una variable numérica.


### **Objetivo Analítico:**

Predecir la satisfacción del cliente ("satisfaction") basándote en los datos disponibles.

El gráfico de dispersión muestra un patrón interesante: cuando los vuelos son más cortos (1200 Km), es poco común encontrar demoras o, si las hay, son generalmente mínimas en comparación con los vuelos de mayor distancia. Además, a medida que el retraso se incrementa en los vuelos de mayor distancia, también tiende a aumentar la insatisfacción entre los pasajeros. Esto sugiere que tanto la longitud del vuelo como la magnitud del retraso influyen en los niveles de satisfacción de los pasajeros.

## **Correlación apartir de variable dummies**

**"Departure Delay in Minutes"** y **"Arrival Delay in Minutes"** tienen una correlación fuerte positiva (0.97), lo que indica que los retrasos en la salida y la llegada están altamente correlacionados. Esta ultima no me parece una varible a considerar en el algoritmo de clasificacion.

```python
data = data.drop(columns=["Arrival Delay in Minutes"])
```

```python
data.head(2)
```

```python
X = data.drop(columns=["satisfaction"])
Y = data["satisfaction"]
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression  
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
X = pd.get_dummies(X, columns=['Class'])

# Label Encoding.
label_encoder = LabelEncoder()
for i in ['Gender', 'Customer Type', 'Type of Travel']:
    X[i] = label_encoder.fit_transform(X[i])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
score = cross_val_score(RandomForestClassifier(), X_train, y_train, cv=5)
```

```python
print(f"Random Forest : Mean Accuracy = {score.mean()}")
```

```python
# Make tha random forest model with specific best hyperparameters.
model = RandomForestClassifier()

# Fit the model.
model.fit(X_train, y_train)

# Predict y-predict.
y_pred = model.predict(X_test)

# Evaluate the accuracy of y-predict.
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc:.3f}")
```

```python
# Get feature importances
feature_importance = model.feature_importances_

# Sort feature importance indices
sorted_idx = np.argsort(feature_importance)

# Retrieve feature names
features = X.columns

# Plot feature importance
plt.figure()
plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center', color='skyblue')
plt.yticks(range(len(sorted_idx)), features[sorted_idx])
plt.xlabel('Feature Importance')
plt.title('Feature Importance')
plt.show()
```

```python
from sklearn.metrics import recall_score

# Supongamos que 'dissatisfied' es tu clase positiva
pos_label = 'dissatisfied'

# Calcula el recall para la clase positiva
recall = recall_score(y_test, y_pred, pos_label=pos_label)
print(f'Recall para {pos_label}: {recall:.2f}')

# Calcula el recall para la otra clase
other_label = 'satisfied'
recall_other = recall_score(y_test, y_pred, pos_label=other_label)
print(f'Recall para {other_label}: {recall_other:.2f}')
```
